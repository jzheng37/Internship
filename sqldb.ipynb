Python 与 Mysql 交互

```python
import mysql.connector
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="",
  database="school_db"
)

mycursor = mydb.cursor()

mycursor.execute("SELECT * FROM students")

results = mycursor.fetchall()
for row in results:
    print(row)

mycursor.close()
mydb.close()
```
